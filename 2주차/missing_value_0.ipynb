```python
import pandas as pd
from pandas import DataFrame
```

```python
df_left = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
 'A': ['A0', 'A1', 'A2', 'A3'],
 'B': [0.5, 2.2, 3.6, 0.4]})

```

```python
df_right = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
'C': ['C2', 'C3', 'C4', 'C5'],
'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
df_all = pd.merge(df_left, df_right, how='outer', on='KEY')
```
